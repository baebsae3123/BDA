```python
pop.columns = ['총인구수','세대수','세대당 인구','남자인구수','여자인구수','남녀비율']
pop
```

```python
[pop.index[i] for i in range(len(pop))]
```

```python
pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(pop['총인구수'])
plt.xticks(rotation=90)
plt.show()
```

```python
pop.sort_values('총인구수')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5,3))

pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y ='총인구수',data=pop1)
plt.xticks(rotation=90)
plt.show()
```
